```python
import numpy as np
import nibabel as nib
import os
from glob import glob
import time
```

```python
def crop_3d_n_center_slices(file_in, slices, file_out):
    img_ori = nib.load(file_in)

    data_ori = img_ori.get_fdata()
    a_max = np.amax(data_ori)
    mid = data_ori.shape[2] // 2
    if data_ori.shape[2] < 80:
        lim_inf,lim_sup = mid - 20, mid + 20
    else:
        lim_inf,lim_sup = mid - 40, mid + 40
    data_ori[:,:,:lim_inf] = a_max
    data_ori[:,:,lim_sup:] = a_max
    
    print(file_in.split("/")[-1], data_ori.shape, lim_inf, lim_sup)
    
    crop_img = nib.Nifti1Image(data_ori, img_ori.affine, img_ori.header, img_ori.extra, img_ori.file_map)
    nib.save(crop_img, file_out)
```

```python
def iter_in_path(input_dir, output_dir):
    files = glob("{}/*.nii.gz".format(input_dir))
    names = [f.split("/")[-1] for f in files]
    for file, name in zip(files, names):
        file_out = "{}/80_{}".format(output_dir, name)
#         print(file, file_out)
        crop_3d_n_center_slices(file, 80, file_out)
```
