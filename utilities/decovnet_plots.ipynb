```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib

acc_text = 'accuracy'

def plot_results(df, df2, title):
#     e = df['epoch']
    e = df['epoch']
    acc = df[acc_text]
    val_acc = df['val_'+acc_text]
#     val_auc = df['val_auc']

    plt.plot(e, acc, 'b', label='Training acc')
    plt.plot(e, val_acc, 'r', label='Validation acc')
    plt.title('Training and Validation accuracy')
    plt.legend()
    
    plt.show()

def plot_ss(df, df2, title):
    plt.ylim(0, 100)
#     plt.ylim(0, 1.1)
    plt.title(title)
    sns.lineplot(data=df2, x="epoch", y=acc_text, legend='brief', label='Acc')
    if df:
        sns.lineplot(data=df, x="epoch", y=acc_text, legend='brief', label='Acc_over')
    sns.lineplot(data=df2, x="epoch", y="val_"+acc_text, legend='brief', label='Val_acc')
    if df:
        sns.lineplot(data=df, x="epoch", y="val_"+acc_text, legend='brief', label='Val_acc_over')
    plt.show()
```
