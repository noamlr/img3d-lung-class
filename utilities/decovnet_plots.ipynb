```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib

def plot_results(df, title):
    e = df['epoch']
    acc = df['accuracy']
    val_acc = df['val_accuracy']
    val_auc = df['val_auc']

    plt.plot(e, acc, 'b', label='Training acc')
    plt.plot(e, val_acc, 'r', label='Validation acc')
    plt.title('Training and Validation accuracy')
    plt.legend()
    
    plt.show()

def plot_ss(df, title):
#     plt.ylim(0.5, 1.1)
    plt.ylim(0, 100)
    plt.title(title)
    sns.lineplot(data=df, x="epoch", y="accuracy", legend='brief', label='Acc')
    sns.lineplot(data=df, x="epoch", y="val_accuracy", legend='brief', label='Val_acc')
    plt.show()
```

```python

df = pd.read_csv('../csv/output/decovnet/decovnet-model/fold6.csv')
# df = pd.read_csv('../models/spgc-ternario-42img/resnet101/fold/axis1/history.csv')
# df = pd.read_csv('../models/spgc-binario/resnet101/fold/axis2/history.csv')
# plot_results(df, '')
plot_ss(df, 'DeCovNet (dados: SPGC-BinÃ¡rio)')
print(df[df['val_accuracy']==max(df['val_accuracy'])])
```
